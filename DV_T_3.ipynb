```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import matplotlib.ticker as mtick
from statsmodels.graphics.mosaicplot import mosaic
# Replace with your actual file path
df = pd.read_csv("/content/employee_records_no_name.csv")
df.head()
ct = pd.crosstab(df['Department'], df['Attrition'])

ct.plot(kind='bar', stacked=True, figsize=(10,6), colormap='viridis')
plt.title('Stacked Bar Chart: Department vs Attrition')
plt.xlabel('Department')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Attrition')
plt.tight_layout()
plt.show()
proportions = ct.div(ct.sum(axis=1), axis=0)
ct.plot(kind='bar', stacked=False, figsize=(10,6), colormap='Set2')
plt.title('Grouped Bar Chart: Department vs Attrition')
plt.xlabel('Department')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Attrition')
plt.tight_layout()
plt.show()


proportions.plot(kind='bar', stacked=True, figsize=(10,6), colormap='coolwarm')
plt.title('Segmented Bar Chart (Proportions): Department vs Attrition')
plt.ylabel('Proportion')
plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter(1.0))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
mosaic(df, ['Department', 'Attrition'], title='Mosaic Plot: Department vs Attrition')
plt.show()
sns.lmplot(data=df, x='Age', y='MonthlyIncome', hue='Gender', aspect=1.5)
plt.title('Scatterplot with Regression Line: Age vs MonthlyIncome')
plt.show()
df.groupby('Department')['MonthlyIncome'].mean().plot(kind='bar', figsize=(8,5), color='skyblue')
plt.title('Average Monthly Income per Department')
plt.ylabel('Average Income')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
plt.figure(figsize=(10,6))
for dept in df['Department'].unique():sns.kdeplot(data=df[df['Department'] == dept], x='MonthlyIncome', label=dept)

plt.title('Grouped Kernel Density: Monthly Income by Department')
plt.legend()
plt.show()
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Department', y='MonthlyIncome', hue='Gender')
plt.title('Box Plot: Monthly Income by Department and Gender')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
plt.figure(figsize=(10,6))
sns.violinplot(data=df, x='JobLevel', y='MonthlyIncome', hue='Attrition', split=True)
plt.title('Violin Plot: Monthly Income by JobLevel and Attrition')
plt.show()
```
